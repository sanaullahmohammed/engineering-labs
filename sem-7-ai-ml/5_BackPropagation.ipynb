```python
import numpy as np

def sigmoid(x):
    return ( 1 / ( 1 + np.exp(-x) ) )

def sigmoid_grad(x):
    return ( x * ( 1 - x ) )

if __name__ == "__main__":
    # Initialization
    X = np.array([ [2,9], [3,6], [4,8] ])
    y = np.array([ [92], [86], [89] ])
    X = X/np.argmax(X)
    y = y/100
    
    # Variables
    epoch = 1000
    eta = 0.1
    ipNeurons = 2
    hNeurons = 3
    opNeurons = 1
    
    # Make CNN
    wh = np.random.uniform(size=(ipNeurons,hNeurons))
    bh = np.random.uniform(size=(1,hNeurons))
    wo = np.random.uniform(size=(hNeurons,opNeurons))
    bo = np.random.uniform(size=(1,opNeurons))
    
    for _ in range(epoch):
        
        # FORWARD PROPAGATION:
        # Input -> Hidden
        h_ip = np.dot(X, wh) + bh
        h_act = sigmoid(h_ip)
        h_grad = sigmoid_grad(h_act)
        
        # hidden -> Output
        o_ip = np.dot(h_act, wo) + bo
        o_act = sigmoid(o_ip)
        o_grad = sigmoid_grad(o_act)
        
        # BACKWARD PROPAGATION
        # Output -> Hidden
        Eo = y - o_act
        d_op = Eo * o_grad
        
        # Hidden -> Input
        Eh = np.dot(d_op,wo.T)
        d_h = Eh * h_grad
        
        # rectification of weights
        wo += np.dot(h_act.T,d_op) * eta
        wh += np.dot(X.T, d_h) * eta
    print(o_act)
```
