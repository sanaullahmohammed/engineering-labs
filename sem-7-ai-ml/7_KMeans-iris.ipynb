```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def kmeans(f1,f2):
    model = KMeans(3).fit(f1).predict(f1)
    plt.scatter(f1[2],f1[3],c=model)
    plt.show()

def gaussianMix(f1,f2):
    model = GaussianMixture(3).fit(f1).predict(f1)
    plt.scatter(f1[2],f1[3],c=model)
    plt.show()

if __name__ == "__main__":
    dataset = load_iris()
    f1 = pd.DataFrame(dataset.data)
    f2 = pd.DataFrame(dataset.target)
    plt.scatter(f1[2], f1[3], c="black")
    plt.show()
    kmeans(f1,f2)
    gaussianMix(f1,f2)
```
