```python
# import stuffs
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

datasets = load_iris()

X = pd.DataFrame(datasets.data)
Y = pd.DataFrame(datasets.target)

# KMeans
Kmeans = KMeans(3).fit(X).predict(X)
plt.title("Hi baby kanda bear")
plt.scatter(X[2], X[3], c=Kmeans)
plt.show()

# Expectation maximization
GMM = GaussianMixture(3).fit(X).predict(X)
plt.title("Hello kanda!")
plt.scatter(X[2], X[3], c=GMM)
plt.show()
```
