```python
import csv
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

data = pd.read_csv('../datasets/naivedata.csv')

X = np.array(data.iloc[:,0:-1])
y = np.array(data.iloc[:,-1])


print(f"OG X: {len(X)} and OG y: {len(y)}")

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2) # X={Attr1,Attr2....,Attr8}; Y={Attr9}

print(f"\nlengthOf(X_train): {len(X_train)} and lengthOf(y_train): {len(y_train)} total 80%")
print(f"\nlengthOf(X_test): {len(X_test)} and lengthOf(y_test): {len(y_test)} total 20%")

model = GaussianNB()

model.fit(X_train, y_train)

prediction = model.predict(X_test)

print(f"\nAccuracy of the algorithm is: {accuracy_score(prediction,y_test)}") # No of coinciding values in Attr-9 the predicted values and random 20% Attr-9 values.
```
