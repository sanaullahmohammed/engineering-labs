```python
import numpy as np
import pandas as pd

data = pd.read_csv('../datasets/candidate.csv')

# selecting all columns except the last one
concepts = np.array(data.iloc[:,0:-1])
print(f"Concepts: {concepts}")

# selecting just the last column(deciding)
target = np.array(data.iloc[:,-1])
print(f"Target: {target}\n\n")

def learn(concepts, target):
    specific_h = concepts[0].copy()
    print("Initialization of specific_h and general_h")
    print(f"Specific Hypothesies: {specific_h}")
    general_h = [["?" for i in range(len(specific_h))] for i in range(len(specific_h))]
    print(f"General Hypothesies: {general_h}")
    
    # used to identify redundant lists inside the general_h
    comparator_h = ['?' for i in range(len(specific_h))]
    
    for i,h in enumerate(concepts):
        print("\n\nFor Loop Starts")
        if target[i] == "Yes":
            print("If instance is Positive: ")
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    specific_h[x] = '?'
        elif target[i] == "No":
            print("If instance is Negative: ")
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    general_h[x][x] = specific_h[x]
        print("Steps of Candidate Elimination Algorithm ",i+1)
        print(specific_h)
        print(general_h)
    
    # remove redundant question tags
    #indices = [i for i,val in enumerate(general_h) if val == comparator_h]
    #for i in indices:
    #    general_h.remove(comparator_h)
    
    while comparator_h in general_h:
        general_h.remove(comparator_h)
    
    return specific_h, general_h

s_final, g_final = learn(concepts, target)
print(f"\n\nFinal Specific_h: {s_final}")
print(f"Final General_h: {g_final}")
```
