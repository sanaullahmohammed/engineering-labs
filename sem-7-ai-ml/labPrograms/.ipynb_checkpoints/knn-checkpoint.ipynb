```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris

iris = load_iris()

print("Iris Data set Loaded...\n")

x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target)

classifier = KNeighborsClassifier(3).fit(x_train, y_train) # trains a knn model where k=3 neighbors
y_pred = classifier.predict(x_test) # y_pred contains predicted y_values using trained KNN model

print("Results of Classification using KNN with K=3\n")

for r in range(0, len(x_test)):
    print(f"Sample: {x_test[r]}, Actual-Label: {y_test[r]}, Predicted-label: {y_pred[r]}") # contrasts difference between prediction and actual
    
print(f"\nClassification Accuracy: {classifier.score(x_test, y_test)}") # gives accuracy in the model's prediction
```
