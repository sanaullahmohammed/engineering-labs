```python
import numpy as np
import pandas as pd

def entropy(target_col):
    val,counts = np.unique(target_col,return_counts = True)
    ent = sum((-counts[i]/np.sum(counts)) * np.log2( counts[i]/np.sum(counts)) for i in range(len(val)))
    return ent

def infoGain(data,features,target):
    te = entropy(data[target])
    val,counts = np.unique(data[features],return_counts = True)
    eg = sum((counts[i]/np.sum(counts)) * entropy(data[data[features] == val[i]][target]) for i in range(len(val)))
    InfoGain = te-eg
    return InfoGain

def ID3(data,features,target):
    if len(np.unique(data[target])) == 1:
        return np.unique(data[target])[0]
    elif len(features) == 0:
        return None
    else:
        # list of gains of all features
        IG = [infoGain(data,f,target) for f in features]
        print(IG)
        # most probable path aka maximum gain
        index = np.argmax(IG)
        
        # selecting attribute with the largest GainValue, firstIterVal: col = 'Outlook'
        col = features[index]
        
        # initializing a dictionary with col as firstKey, firstIterVal: tree = {'Outlook':{}}
        tree = {col:{}}
        
        # remaining attributes excluding the maxGain attribute, firstIterVal: features = [Temperature, Humidity,  Windy]
        features = [f for f in features if f!=col]
        
        for val in np.unique(data[col]):
            # print(f"Inside forLoop-> sub_data: {data[data[col]==val]};")
            
            # collects records where colValues are similar to val, stores them into sub_data, check 2nd condition of ID3 function to understand dropna() 
            sub_data = data[data[col]==val].dropna()
            print(f"SUBDATA: {sub_data}")
            
            # subTree returns Yes or No values, check the first condition in ID3 function
            subtree = ID3(sub_data,features,target)
            
            # recursively filling the dictionary values
            tree[col][val] = subtree
        return tree
        
        '''
        Outlook Temperature Humidity  Windy PlayTennis
        0   Sunny         Hot     High  False         No
        1   Sunny         Hot     High   True         No
        7   Sunny        Mild     High  False         No
        '''

data = pd.read_csv('../datasets/PlayTennis.csv')
testData = data.sample(frac = 0.1)
print(testData)
target = 'PlayTennis'
features = data.columns[data.columns!=target]
tree = ID3(data,features,target)
print(tree)
```

```python
import numpy as np
import pandas as pd

def entropy(target_col):
    val,counts = np.unique(target_col,return_counts = True)
    ent = sum( (-counts[i]/np.sum(counts)) * np.log2( counts[i]/np.sum(counts) ) for i in range(len(val)))
    return ent

def infoGain(data,features,target):
    te = entropy(data[target])
    val,counts = np.unique(data[features],return_counts = True)
    eg = sum((counts[i]/sum(counts)) * entropy(data[data[features] == val[i]][target] ) for i in range(len(val))) 
    return te-eg

def ID3(data,features,target):
    if len(np.unique(data[target])) == 1:
        return np.unique(data[target])[0]
    elif len(features) == 0:
        return None
    else:
        IG = [infoGain(data,f,target) for f in features]
        index = np.argmax(IG)
        col = features[index]
        tree = {col:{}}
        features = [f for f in features if f!=col]
        for val in np.unique(data[col]):
            sub_data = data[data[col]==val].dropna()
            subtree = ID3(sub_data,features,target)
            tree[col][val] = subtree
        return tree


data = pd.read_csv('../datasets/PlayTennis.csv')
target = 'PlayTennis'
features = data.columns[data.columns!=target]
tree = ID3(data,features,target)
print (tree)
```
