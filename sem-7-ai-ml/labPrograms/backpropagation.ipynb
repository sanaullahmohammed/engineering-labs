```python
import numpy as np

X = np.array([[2,9], [3,6], [4,8]])
Y = np.array([[92], [86], [89]])
X = X/np.amax(X,axis=0) # axis0 : columnWise ; axis1 : rowWise
Y = Y/100 # normalizing outputs to range from 0,1

def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_grad(x):
    return x * (1-x)

epoch = 1000 # Number of iterations
eta = 0.1 # learning rate
input_neurons = 2 # 2 input neurons (1 layer of inputs)
hidden_neurons = 3 # 3 hidden neurons (1 layer of hidden)
output_neurons = 1 # 1 output neurons (1 layer of outputs)

wh = np.random.uniform(size = (input_neurons, hidden_neurons)) # weights between 2 i/p and 3 hidden = 6 weights
bh = np.random.uniform(size = (1, hidden_neurons)) # bias for each hidden neuron
wout = np.random.uniform(size = (hidden_neurons, output_neurons)) # weights between 3 hidden and 1 output = 3 weights
bout = np.random.uniform(size = (1, output_neurons)) # bias for output neuron

for i in range(epoch):
    
    # np.dot == Matrix-Multiplication
    # ".T" operator is Transpose operator
    
    # feed forward Network
    h_ip = np.dot(X, wh) + bh # array of hiddenNeurons(H(i)| i={1,2,3}) using formula: Hi = Sum(W.x)+bias(i)
    h_act = sigmoid(h_ip) # activation function of the hiddenNeuron h_ip; (Sigmoid(h_ip))
    o_ip = np.dot(h_act, wout) + bout # array of outputNeuron(H(i)| i={1}) using formula: Hi = Sum(W.x)+bias(i)
    output = sigmoid(o_ip) # activation function of the outputNeuron o_ip; (Sigmoid(o_ip))
    
    # back propagate (RHS of HiddenLayer aka, HiddenLayer, Output)
    Eo = Y - output # difference between expected o/p and actual o/p
    outgrad = sigmoid_grad(output) # differentialSigmoid function
    d_output = Eo * outgrad # error differential
    Eh = np.dot(d_output, wout.T) # rectifying hidden->output weights
    
    # back propagate (LHS of HiddenLayer aka, HiddenLayer, Input)
    hiddengrad = sigmoid_grad(h_act)
    d_hidden = Eh * hiddengrad
    
    # rectifying Hidden -> Output weights
    wout += np.dot(h_act.T, d_output) * eta
    
    # rectifying Input -> Hidden weights
    wh += np.dot(X.T, d_hidden) * eta
    
print("Normalized Input: \n", X)
print("Actual Output: \n", Y)
print("Predicted Output: \n", output)
```

```python
import numpy as np
#numpy is commonly used to process number array
X = np.array ([[2, 9],[3, 6],[4, 8]])
#Features ( Hrs Slept, Hrs Studied)
y = np.array ([[92],[86],[89]])
#Labels(Marks obtained)
X = X / np.amax (X, axis = 0)
#Normalize
y = y / 100 
def sigmoid (x): 
    return 1 / (1 + np.exp (-x)) 
def sigmoid_grad (x):
    return x * (1 -x)
  
#Variable initialization
epoch = 1000
#Setting training iterations
eta = 0.1
#Setting learning rate (eta)
input_neurons = 2
#number of features in data set
hidden_neurons = 3
#number of hidden layers neurons
output_neurons = 1
#number of neurons at output layer
#Weight and bias - Random initialization
wh = np.random.uniform (size = (input_neurons, hidden_neurons))
#1x3
bh = np.random.uniform (size = (1, hidden_neurons))
#3x1
wout = np.random.uniform (size = (hidden_neurons, output_neurons))
#1x1
bout = np.random.uniform (size = (1, output_neurons)) 
for i in range (epoch):
#Forward Propogation
    h_ip = np.dot (X, wh) + bh
#Dot product + bias
    h_act = sigmoid (h_ip)
#Activation function
    o_ip = np.dot (h_act, wout) + bout 
    output = sigmoid (o_ip)
#Error at Output layer
    Eo = y - output
#Error at o/p
    outgrad = sigmoid_grad (output) 
    d_output = Eo * outgrad
#Errj=Oj(1-Oj)(Tj-Oj)
#Error at Hidden later
    Eh = np.dot (d_output, wout.T)
#.T means transpose
    hiddengrad = sigmoid_grad (h_act)
#How much hidden layer wts contributed to error
    d_hidden = Eh * hiddengrad 
    wout += np.dot (h_act.T, d_output) * eta
#Dotproduct of nextlayererror and currentlayerop
    wh += np.dot (X.T, d_hidden) * eta 
print ("Normalized Input: \n", X) 
print ("Actual Output: \n", y) 
print ("Predicted Output: \n", output)
```
