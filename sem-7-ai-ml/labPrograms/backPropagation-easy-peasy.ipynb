```python
#numpy is commonly used to process number array
import numpy as np

#Features ( Hrs Slept, Hrs Studied)
X = np.array ([[2, 9],[3, 6],[4, 8]])

#Labels(Marks obtained)
y = np.array ([[92],[86],[89]])

#Normalize (np.amax => [4,9])
X = X / np.amax (X, axis = 0)
y = y / 100

def sigmoid (x):
    return 1 / (1 + np.exp (-x)) 

def sigmoid_grad (x):
    return x * (1 -x)
  
#Variable initialization

#Setting training iterations
epoch = 1000

#Setting learning rate (eta)
eta = 0.1

#number of features in data set
input_neurons = 2

#number of hidden layers neurons
hidden_neurons = 3

#number of neurons at output layer
output_neurons = 1

#Weight and bias - Random initialization

#2x3
wh = np.random.uniform (size = (input_neurons, hidden_neurons))

#1x3
bh = np.random.uniform (size = (1, hidden_neurons))

#3x1
wout = np.random.uniform (size = (hidden_neurons, output_neurons))

#1x1
bout = np.random.uniform (size = (1, output_neurons)) 

for i in range (epoch):
    #Forward Propogation
    
    #Dot product + bias; X(3,2) <-> wh(2,3) <-> bh(3,1)
    h_ip = np.dot (X, wh) + bh

    #Activation function
    h_act = sigmoid (h_ip)
    o_ip = np.dot (h_act, wout) + bout 

    #Error at Output layer
    output = sigmoid (o_ip)

    #Error at o/p
    Eo = y - output
    outgrad = sigmoid_grad(output) 

    #Errj=Oj(1-Oj)(Tj-Oj)
    #Error at Hidden later
    d_output = Eo * outgrad

    #.T means transpose
    Eh = np.dot (d_output, wout.T)

    #How much hidden layer wts contributed to error
    hiddengrad = sigmoid_grad (h_act)

    d_hidden = Eh * hiddengrad 
    
    #Dotproduct of nextlayererror and currentlayerop
    wout += np.dot (h_act.T, d_output) * eta

    wh += np.dot (X.T, d_hidden) * eta 

print ("Normalized Input: \n", X) 
print ("Actual Output: \n", y) 
print ("Predicted Output: \n", output)
```
