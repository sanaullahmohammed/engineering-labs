```python
def aStarAlgo(start_node, stop_node):
    # Initializing Variables
    open_set = set(start_node) # tobe Explored
    closed_set = set() # already explored
    g = {} # distance from starting node
    parents = {}
    g[start_node] = 0
    parents[start_node] = start_node
    
    while len(open_set) > 0:
        n = None # ideal vertex
        
        # loop to find ideal(least) value vertex
        for v in open_set:
            if n == None or g[v] + H_dist[v] < g[n] + H_dist[n]:
                n = v
        
        if n == None:
            print("Path doesn't exist!")
            return None
        
        if n == stop_node:
            reconst_path = []
            
            while parents[n] != n:
                reconst_path.append(n)
                n = parents[n]
            
            reconst_path.append(start_node)
            reconst_path.reverse()
            print(f"Path found: {reconst_path}")
            return reconst_path
        
        # check if goalNode is reached or No nodes remain
        if n != stop_node or Graph_nodes[n] != None:
            for (m,weight) in get_neighbors(n):
                
                # vertex visited first time
                if m not in open_set and m not in closed_set:
                    open_set.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight
                
                elif g[m] > g[n] + weight:
                    g[m] = g[n] + weight
                    parents[m] = n
                    if m in closed_set:
                        closed_set.remove(m)
                        open_set.add(m)
                    
        open_set.remove(n)
        closed_set.add(n)
    
    print("Path doesn't exist!")
    return None
```

```python
def get_neighbors(v):
    if v in Graph_nodes:
        return Graph_nodes[v]
    else:
        return None
```

```python
Graph_nodes = {
    'A': [('B', 2), ('E', 3)],
    'B': [('C', 1), ('G', 9)],
    'C': None,
    'E': [('D', 6)],
    'D': [('G', 1)],
}
```

```python
H_dist = {
    'A': 11,
    'B': 6,
    'C': 99,
    'D': 1,
    'E': 7,
    'G': 0,
}
```
