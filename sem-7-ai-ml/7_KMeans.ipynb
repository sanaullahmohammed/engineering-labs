```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def kmeans(f1,f2,X):
    model = KMeans(3).fit(X).predict(X)
    plt.scatter(f1,f2,c=model)
    plt.show()

def gaussianMix(f1, f2, X):
    model = GaussianMixture(3).fit(X).predict(X)
    plt.scatter(f1,f2,c=model)
    plt.show()

if __name__ == "__main__":
    data = pd.read_csv("../datasets/8-kmeans.csv")
    f1 = data["Distance_Feature"]
    f2 = data["Speeding_Feature"]
    X = np.array(list(zip(f1,f2)))
    plt.scatter(f1,f2,c="black")
    plt.show()
    kmeans(f1,f2,X)
    gaussianMix(f1,f2,X)
```
