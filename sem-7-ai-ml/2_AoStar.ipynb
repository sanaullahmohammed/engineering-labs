```python
class Graph:
    def __init__(self, graph, heuristic, start):
        self.graph = graph
        self.H = heuristic
        self.start = start
        self.parent = dict()
        self.status = dict()
        self.solutionGraph = dict()
        
    def getNeighbors(self, v):
        return self.graph.get(v,'')

    def getStatus(self,v):
        return self.status.get(v,0)

    def setStatus(self,v, val): 
        self.status[v]=val

    def getHeuristic(self, n):
        return self.H.get(n,0) 

    def setHeuristic(self, n, value):
        self.H[n]=value 

    #def printSolution(self):
        #print("GRAPH SOLUTION,",self.start)
        #print("------------------------------------------------------------")
        #print(self.solutionGraph)
        #print("------------------------------------------------------------")

    def minimumCost(self, v): 
        minCost=0
        costToNode={}
        costToNode[minCost]=[]
        flag=True
        for nodeTupleList in self.getNeighbors(v): 
            cost=0
            nodeList=[]
            for c, weight in nodeTupleList:
                cost += self.getHeuristic(c)+weight
                nodeList.append(c)
            if flag==True: 
                minCost=cost
                costToNode[minCost]=nodeList 
                flag=False
            elif minCost>cost:
                minCost=cost
                costToNode[minCost]=nodeList 
        return minCost, costToNode[minCost] 

    def aoStar(self, v, backTracking): 
        #print("HEURISTIC VALUES :", self.H)
        #print("SOLUTION GRAPH :", self.solutionGraph)
        #print("PROCESSING NODE :", v)
        #print("-----------------------------------------------------------------------------------------")
        if self.getStatus(v) >= 0: 
            minCost, NodeList = self.minimumCost(v)
            #print(minCost, NodeList)
            self.setHeuristic(v, minCost)
            self.setStatus(v,len(NodeList))
            solved=True # check the Minimum Cost nodes of v are solved
            for node in NodeList:
                self.parent[node]=v
                #if self.getStatus(node)!=-1:
                #    solved=solved & False
            if solved==True: 
                self.setStatus(v,-1)
                self.solutionGraph[v]=NodeList 
            if v!=self.start:
                self.aoStar(self.parent[v], True) 
            if backTracking==False: # check the current call is not for backtracking 
                for node in NodeList:
                    self.setStatus(node,0) 
                    self.aoStar(node, False)  


#for simplicity we ll consider heuristic distances given
#print ("Graph - 1")
h1 = {'A': 1, 'B': 6, 'C': 2, 'D': 12, 'E': 2, 'F': 1, 'G': 5, 'H': 7, 'I': 7, 'J': 1}
graph1 = {
    'A': [[('B', 1), ('C', 1)], [('D', 1)]],
    'B': [[('G', 1)], [('H', 1)]],
    'C': [[('J', 1)]],
    'D': [[('E', 1), ('F', 1)]],
    'G': [[('I', 1)]]
}

G1= Graph(graph1, h1, 'A')
G1.aoStar('A',False)
#G1.printSolution()
print(G1.solutionGraph)
```
