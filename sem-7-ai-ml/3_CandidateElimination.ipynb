```python
import pandas as pd
import numpy as np

def learn(concepts, target):
    specific = concepts[0].copy()
    general = [ ['?' for _ in range(len(specific))] for _ in range(len(specific)) ]
    comparator = ['?' for _ in range(len(specific))]
    
    for i,h in enumerate(concepts):
        if target[i] == "Yes":
            for x in range(len(specific)):
                if h[x] != specific[x]:
                    specific[x] = '?'
        else:
            for x in range(len(specific)):
                if h[x] != specific[x]:
                    general[x][x] = specific[x]
    
    while comparator in general:
        general.remove(comparator)
    return specific, general

if __name__ == "__main__":
    data = pd.read_csv("../datasets/candidate.csv")
    concepts = np.array(data.iloc[:,:-1])
    target = np.array(data.iloc[:,-1])
    specific, general = learn(concepts, target)
    print(f"Specific: {specific}\nGeneral: {general}")
```
